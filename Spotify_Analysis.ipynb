## Exploratory Data Analysis of Spotify Dataset

## Importing the modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn
```

## Importing the Data

## Converting the duration from milliseconds to seconds

```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
df_tracks.head()
```

## Finding the 10 least popular songs

```python
sorted_df=df_tracks.sort_values('popularity',ascending=True)
sorted_df.head(10)
```

## Finding the 10 most popular songs 

```python
sorted_df=df_tracks.sort_values('popularity',ascending=True)
sorted_df.tail(10)
```

## Most Popular songs that are greater than 90

```python
most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=True)
most_popular.head(10)
```

## Changing the index of the required coloumns

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

## Removing the unwanted columns 

```python
corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
corr_df.head()
```

## Creating a co-relation map

```python
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g',cmap='inferno',linecolor='black' )
heatmap.set_title('Correlation Heatmap between variable')
heatmap.set_xticklabels(heatmap.get_xticklabels())
plt.show()
```

## Creating a sample DataFrame

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
sample_df.shape
```

## Plotting a Regression Graph between Energy and Loudness

```python
plt.figure(figsize=(10,5))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c')
plt.show()
```

## Plotting a Regression Graph between Aucosticness and Popularity 

```python
plt.figure(figsize=(10,5 ))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='m')
plt.show()
```

## Plotting a Histogram for Songs based on the Release Year

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.danceability=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of Songs released every Year')
plt.show()
```

## Plotting a graph for Duration trend over the years

```python
total_duration=df_tracks.duration
fig_dims=(18,7)
fig, ax = plt.subplots(figsize = (20,7))
fig=sns.barplot(x=years, y=total_duration, ax=ax, errwidth=False).set(title='Duration trend over the Years')
plt.xticks(rotation=90)
plt.show()
```

## Plotting a graph for Average Duration over the years

```python
total_duration=df_tracks.duration
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig=sns.lineplot(x=years, y=total_duration).set(title='Average duration over the Years')
plt.xticks(rotation=90)
plt.show()
```
